```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta

start_date = datetime(2023, 8, 15)  # Using an arbitrary year
new_df['adjusted_day'] = new_df['Date of Injury'].apply(lambda x: (x - start_date).days % 365)

injury_counts = new_df['Sorted Injuries'].value_counts(ascending=False)
ordered_injury_types = injury_counts.index[::-1]

palette_colors = sns.color_palette('tab20', n_colors=len(ordered_injury_types))
color_map = dict(zip(ordered_injury_types, palette_colors))

sns.set_theme(style="ticks")
f, ax = plt.subplots(figsize=(15, 10))
sns.despine(f)

sns.histplot(
    data=new_df,
    x="adjusted_day",
    hue="Sorted Injuries",
    hue_order=ordered_injury_types,
    multiple="stack",
    palette=color_map,
    edgecolor=".3",
    linewidth=.5,
    bins=13  # 52 weeks / 4 = 13 bins for 4-week periods
)

def format_x_label(x, pos):
    date = start_date + timedelta(days=int(x))
    return date.strftime('%d %b')

ax.xaxis.set_major_formatter(plt.FuncFormatter(format_x_label))
ax.set_xticks(range(0, 365, 28))  # Set ticks every 4 weeks
ax.set_xlabel("Date", fontsize=12)
ax.set_xlim(0, 365)  # Set the x-axis limits to start from 0
ax.set_ylabel("Count of Injuries", fontsize=12)
plt.title("Injury Distribution by 4-Week Periods (Starting August 15th)", fontsize=14, y=1.05)

legend_labels = [f"{injury} ({count})" for injury, count in injury_counts.items()]
plt.legend(title="Injury Types", labels=legend_labels, loc='upper right', bbox_to_anchor=(1.45, 1))

ax.tick_params(axis='both', which='major', labelsize=10)
ax.grid(axis='y', linestyle=':', linewidth=0.5, color='gray', alpha=0.5)

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
injuries_per_month = new_df.groupby("month_number")["Sorted Injuries"].count().reset_index(name="Injury Count").sort_values(by="month_number")
median_offpitch_per_month = new_df.groupby("month_number")["days off the pitch"].median().reset_index(name="Median Offpitch")
injuries_per_month = pd.merge(injuries_per_month, median_offpitch_per_month, on="month_number", how="left")

print("Injury Count by Month:\n", injuries_per_month.to_string(index=False))

offpitch_stats = new_df.groupby("Sorted Injuries")["days off the pitch"].agg(["mean", "median", "std", "min", "max"])
print("\nDays off the pitch Statistics by Sorted Injuries:\n", offpitch_stats)
```

```python
start_date = 152  # June 1
end_date = 236    # August 24

nosummer_df = new_df[~new_df['day_of_year'].between(start_date, end_date)]
nosummer_df.sample(5)
```

```python
s_injuries_by_month = nosummer_df.groupby("month_number")["Sorted Injuries"].count().reset_index(name="injury_count").sort_values(by="month_number")
s_median_offpitch_by_month = nosummer_df.groupby("month_number")["days off the pitch"].median().reset_index(name="median_offpitch")
s_injuries_by_month = pd.merge(s_injuries_by_month, s_median_offpitch_by_month, on="month_number", how="left")

print("Injury Count by Month:\n", s_injuries_by_month.to_string(index=False))

s_offpitch_stats_by_injury = nosummer_df.groupby("Sorted Injuries")["days off the pitch"].agg(["mean", "median", "std", "min", "max"])
print("\nOffpitch Statistics by Sorted Injuries:\n", s_offpitch_stats_by_injury)
```

```python
combined_stats = pd.concat(
    [
        nosummer_df.groupby("Sorted Injuries")["days off the pitch"].agg(["mean", "median", "std", "count"]),
        new_df.groupby("Sorted Injuries")["days off the pitch"].agg(["mean", "median", "std", "count"])
    ],
    axis=1
)

# Round all columns to 0 decimal places
combined_stats = combined_stats.round(0)

combined_stats.columns = [
    "wo_mean", "wo_median", "wo_std", "wo_count", 
    "full_mean", "full_median", "full_std", "full_count"
]

# Sort by full_count in descending order
combined_stats.sort_values("full_count", ascending=False)
```
