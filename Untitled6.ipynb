```python
import pandas as pd
import matplotlib.pyplot as plt

# Load datasets
df1 = pd.read_csv(r"C:\Users\HP\Downloads\injury_risk_dataset_v2.csv")

# Basic info about the datasets
print("Dataset 1 Shape:", df1.shape)
print("\nDataset 1 Columns:", df1.columns.tolist())
```

```python
missing_values = df1.isnull().sum()
print("\nMissing Values per Column:\n", missing_values)

injury_distribution = df1['Injury'].value_counts()
print("\nInjury Distribution:\n", injury_distribution)

injury_rate = injury_distribution[1] / df1.shape[0] * 100
print(f"\nInjury Rate: {injury_rate:.2f}%")
```

```python
injury_by_position = df1.groupby("Position")["Injury"].mean().sort_values(ascending=False)

plt.figure(figsize=(8, 5))
injury_by_position.plot(kind="bar")
plt.title("Injury Rate by Position")
plt.ylabel("Injury Rate")
plt.xticks(rotation=45)
plt.show()
```

```python
workload_cols = ["Distance_km", "High_Speed_km", "Sprints",
                 "Match_Minutes", "Accelerations", "Decelerations"]

workload_comparison = df1.groupby("Injury")[workload_cols].mean()
print("\nWorkload Comparison:\n", workload_comparison)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

df = df1.copy()

df = pd.get_dummies(df, columns=['Position'], drop_first=True)

df['HighSpeed_Ratio'] = df['High_Speed_km'] / df['Distance_km']
df['Accel_to_Decel_Ratio'] = df['Accelerations'] / (df['Decelerations'] + 1e-6)

df = df.drop(columns=['Player_ID', 'Date', 'Injury_Type', 'Days_Missed'])

X = df.drop(columns=['Injury'])
y = df['Injury']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("Training Set Shape:", X_train.shape)
print("Testing Set Shape:", X_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix, precision_recall_fscore_support

rf_model = RandomForestClassifier(
    n_estimators=200,
    class_weight='balanced',
    random_state=42
)
rf_model.fit(X_train, y_train)

rf_preds = rf_model.predict(X_test)
rf_proba = rf_model.predict_proba(X_test)[:, 1]

print("=== Random Forest Results ===")
print("ROC-AUC:", roc_auc_score(y_test, rf_proba))
print(classification_report(y_test, rf_preds, digits=3))
print("Confusion Matrix:\n", confusion_matrix(y_test, rf_preds))

xgb_model = XGBClassifier(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=5,
    scale_pos_weight=(y_train.value_counts()[0]/y_train.value_counts()[1]), # handle imbalance
    use_label_encoder=False,
    eval_metric='logloss',
    random_state=42
)
xgb_model.fit(X_train, y_train)

xgb_preds = xgb_model.predict(X_test)
xgb_proba = xgb_model.predict_proba(X_test)[:, 1]

print("\n=== XGBoost Results ===")
print("ROC-AUC:", roc_auc_score(y_test, xgb_proba))
print(classification_report(y_test, xgb_preds, digits=3))
print("Confusion Matrix:\n", confusion_matrix(y_test, xgb_preds))

import matplotlib.pyplot as plt
import numpy as np

importances = rf_model.feature_importances_
indices = np.argsort(importances)[::-1]
features = X_train.columns

plt.figure(figsize=(10,5))
plt.bar(range(len(importances)), importances[indices])
plt.xticks(range(len(importances)), features[indices], rotation=90)
plt.title("Random Forest Feature Importance")
plt.show()
```

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import LabelEncoder
import random

positions = ["Goalkeeper", "Defender", "Midfielder", "Forward"]
injury_types = ["None", "Hamstring", "ACL Tear", "Ankle Sprain", "Groin Strain", "Concussion"]
injury_probs = [0.85, 0.05, 0.03, 0.04, 0.02, 0.01]

start_date = datetime(2025, 1, 1)
dates = [start_date + timedelta(days=i) for i in range(1000)]
data = []

for i in range(1000):
    player_id = random.randint(1, 25)
    date = dates[i]
    position = random.choice(positions)
    distance = round(np.random.normal(9, 1.5), 2)
    high_speed = round(np.random.normal(2, 0.5), 2)
    sprints = max(0, int(np.random.normal(12, 3)))
    minutes = random.choice([0, 30, 60, 90])
    accel = max(0, int(np.random.normal(20, 5)))
    decel = max(0, int(np.random.normal(18, 4)))
    hr_zone = round(np.random.uniform(3, 5), 1)
    injury_type = np.random.choice(injury_types, p=injury_probs)
    injury = 0 if injury_type == "None" else 1
    days_missed = 0 if injury == 0 else random.randint(5, 60)

    data.append([
        player_id, date.strftime("%Y-%m-%d"), position,
        distance, high_speed, sprints, minutes,
        accel, decel, hr_zone, injury, injury_type, days_missed
    ])

df = pd.DataFrame(data, columns=[
    "Player_ID", "Date", "Position",
    "Distance_km", "High_Speed_km", "Sprints", "Match_Minutes",
    "Accelerations", "Decelerations", "HR_Zone",
    "Injury", "Injury_Type", "Days_Missed"
])

df['Position'] = LabelEncoder().fit_transform(df['Position'])
df['Injury_Type'] = LabelEncoder().fit_transform(df['Injury_Type'])

X = df.drop(columns=["Date", "Injury", "Injury_Type", "Days_Missed"])
y = df["Injury"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

sm = SMOTE(random_state=42)
X_train_res, y_train_res = sm.fit_resample(X_train, y_train)

rf_model = RandomForestClassifier(class_weight='balanced', random_state=42)
rf_model.fit(X_train_res, y_train_res)
rf_preds = rf_model.predict(X_test)
rf_auc = roc_auc_score(y_test, rf_model.predict_proba(X_test)[:, 1])
print("=== Random Forest ===")
print("ROC-AUC:", rf_auc)
print(classification_report(y_test, rf_preds))
print("Confusion Matrix:\n", confusion_matrix(y_test, rf_preds))

xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
xgb_model.fit(X_train_res, y_train_res)
xgb_preds = xgb_model.predict(X_test)
xgb_auc = roc_auc_score(y_test, xgb_model.predict_proba(X_test)[:, 1])
print("\n=== XGBoost ===")
print("ROC-AUC:", xgb_auc)
print(classification_report(y_test, xgb_preds))
print("Confusion Matrix:\n", confusion_matrix(y_test, xgb_preds))
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, roc_auc_score, precision_recall_curve, confusion_matrix
import numpy as np
import matplotlib.pyplot as plt

smote = SMOTE(random_state=42)
X_train_balanced, y_train_balanced = smote.fit_resample(X_train, y_train)

print("Before SMOTE:", X_train.shape, "Injury=1 count:", sum(y_train))
print("After SMOTE:", X_train_balanced.shape, "Injury=1 count:", sum(y_train_balanced))

rf_balanced = RandomForestClassifier(
    n_estimators=300,
    random_state=42
)
rf_balanced.fit(X_train_balanced, y_train_balanced)

rf_proba_balanced = rf_balanced.predict_proba(X_test)[:, 1]

threshold = 0.25  # can experiment with 0.2–0.3
rf_preds_balanced = (rf_proba_balanced >= threshold).astype(int)

print("=== Random Forest with SMOTE & Threshold 0.25 ===")
print("ROC-AUC:", roc_auc_score(y_test, rf_proba_balanced))
print(classification_report(y_test, rf_preds_balanced, digits=3))
print("Confusion Matrix:\n", confusion_matrix(y_test, rf_preds_balanced))

precision, recall, thresholds = precision_recall_curve(y_test, rf_proba_balanced)
plt.figure(figsize=(6,4))
plt.plot(recall, precision)
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve (Random Forest with SMOTE)")
plt.grid(True)
plt.show()
```
