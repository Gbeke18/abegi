```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score

# Load enhanced dataset
df = pd.read_csv(r"C:\Users\HP\Downloads\injury_risk_enhanced_dataset.csv")

# One-hot encode 'Position'
df = pd.get_dummies(df, columns=['Position'], drop_first=True)

# Define features and target
X = df.drop(columns=['Player_ID', 'Date', 'Injury', 'Injury_Type', 'Days_Missed'])
y = df['Injury']

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)

# Train models
rf = RandomForestClassifier(random_state=42)
xgb = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)

rf.fit(X_train, y_train)
xgb.fit(X_train, y_train)

# Evaluate
for model, name in zip([rf, xgb], ['Random Forest', 'XGBoost']):
    preds = model.predict(X_test)
    proba = model.predict_proba(X_test)[:, 1]
    print(f"\n=== {name} ===")
    print(f"ROC-AUC: {roc_auc_score(y_test, proba):.3f}")
    print(classification_report(y_test, preds))
    print("Confusion Matrix:\n", confusion_matrix(y_test, preds))

from sklearn.model_selection import RandomizedSearchCV

param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [5, 10, 15, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

search = RandomizedSearchCV(rf, param_grid, cv=3, n_iter=10, scoring='roc_auc', random_state=42)
search.fit(X_train, y_train)
best_rf = search.best_estimator_

print("Best Random Forest ROC-AUC:", roc_auc_score(y_test, best_rf.predict_proba(X_test)[:, 1]))

import shap

# Create TreeExplainer
explainer = shap.TreeExplainer(xgb)
shap_values = explainer.shap_values(X_test)

# Summary Plot
shap.summary_plot(shap_values, X_test, plot_type="bar")

# Example: sort and group
df['Date'] = pd.to_datetime(df['Date'])
df.sort_values(by=['Player_ID', 'Date'], inplace=True)

# Create sequences (simplified version)
sequence_length = 5
sequences = []
labels = []

players = df['Player_ID'].unique()

for player in players:
    group = df[df['Player_ID'] == player]
    group = group.reset_index(drop=True)
    for i in range(len(group) - sequence_length):
        seq = group.iloc[i:i+sequence_length][X.columns].values
        label = group.iloc[i+sequence_length]['Injury']
        sequences.append(seq)
        labels.append(label)

# Convert to numpy arrays
import numpy as np
X_seq = np.array(sequences)
y_seq = np.array(labels)
```
