```python
# -*- coding: utf-8 -*-
"""
Advanced Football Injury Prediction System
Time Series Analysis with External Factors
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
import warnings
warnings.filterwarnings('ignore')

# Set style for visualizations
plt.style.use('ggplot')
sns.set_palette("husl")
```

```python
# Load your dataset
df = pd.read_csv(r"C:\Users\HP\Downloads\data.csv")

# Display basic info
print("Dataset Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nMissing Values:\n", df.isnull().sum())
print("\nInjury Distribution:\n", df['Injury_Next_Season'].value_counts())
```

```python
# Original feature engineering
df['Workload_Stress_Index'] = df['Training_Hours_Per_Week'] * df['Stress_Level_Score']
df['Sleep_Efficiency'] = df['Sleep_Hours_Per_Night'] / df['Training_Hours_Per_Week']
df['Recovery_Score'] = (
    0.4 * df['Sleep_Hours_Per_Night'] +
    0.4 * (df['Nutrition_Quality_Score'] / 100) +
    0.2 * df['Warmup_Routine_Adherence']
)
df['Injury_Risk_Score'] = (
    df['Previous_Injury_Count'] * 2 -
    df['Hamstring_Flexibility'] * 0.1 -
    df['Balance_Test_Score'] * 0.1 +
    df['Stress_Level_Score'] * 0.2
)

# New: Calculate BMI properly
df['BMI'] = df['Weight_kg'] / ((df['Height_cm'] / 100) ** 2)

# New: Fitness to workload ratio
df['Fitness_Workload_Ratio'] = (df['Knee_Strength_Score'] + df['Hamstring_Flexibility']) / df['Training_Hours_Per_Week']

# New: Recovery deficit
df['Recovery_Deficit'] = df['Stress_Level_Score'] - df['Recovery_Score']

# Display new features
df[['Workload_Stress_Index', 'Sleep_Efficiency', 'Recovery_Score', 
    'Injury_Risk_Score', 'BMI', 'Fitness_Workload_Ratio', 'Recovery_Deficit']].head()
```

```python
# Create more realistic time series data
np.random.seed(42)
n_weeks = 12  # Extended to 12 weeks for better time series analysis
players = df.copy()
player_ids = range(len(players))

ts_data = []

for player_id, (_, row) in zip(player_ids, players.iterrows()):
    base_load = row['Training_Hours_Per_Week']
    stress_level = row['Stress_Level_Score'] / 100  # Normalize to 0-1
    
    for week in range(1, n_weeks + 1):
        # More realistic fluctuations based on player profile
        if row['Position'] == 'Forward':
            fluctuation = np.random.normal(loc=1.1, scale=0.15)  # Forwards train more
        elif row['Position'] == 'Defender':
            fluctuation = np.random.normal(loc=0.9, scale=0.1)   # Defenders train less
        else:
            fluctuation = np.random.normal(loc=1.0, scale=0.1)   # Midfielders average
            
        # Add weekly variation (players train less as season progresses)
        season_factor = 1 - (week / n_weeks) * 0.1  # 10% reduction by end of season
        
        # Stress impact on training
        stress_factor = 1 - (stress_level * 0.2)  # High stress reduces training quality
        
        session_load = base_load * fluctuation * season_factor * stress_factor
        
        ts_data.append({
            'player_id': player_id,
            'week': week,
            'session_load': max(1, session_load),  # Ensure positive
            'position': row['Position']
        })

df_sessions = pd.DataFrame(ts_data)
df_sessions.sort_values(by=['player_id', 'week'], inplace=True)

# Calculate advanced time series features
def calculate_time_series_features(group):
    group = group.sort_values('week')
    
    # Acute and chronic loads
    group['acute_load'] = group['session_load'].rolling(window=1, min_periods=1).mean()
    group['chronic_load'] = group['session_load'].rolling(window=4, min_periods=1).mean()
    
    # ACWR - Acute:Chronic Workload Ratio
    group['ACWR'] = group['acute_load'] / group['chronic_load']
    
    # Load trends
    group['load_trend_7d'] = group['session_load'].rolling(window=7, min_periods=1).mean()
    group['load_std_7d'] = group['session_load'].rolling(window=7, min_periods=1).std()
    
    # Change metrics
    group['load_change_pct'] = group['session_load'].pct_change()
    group['load_spike'] = (group['load_change_pct'] > 0.2).astype(int)  # >20% increase
    
    # Monotony (daily load variability)
    group['monotony'] = group['session_load'].rolling(window=7, min_periods=1).mean() / \
                        group['session_load'].rolling(window=7, min_periods=1).std()
    
    # Strain (load √ó monotony)
    group['strain'] = group['session_load'] * group['monotony']
    
    return group

df_sessions = df_sessions.groupby('player_id').apply(calculate_time_series_features)
df_sessions.reset_index(drop=True, inplace=True)

# Handle infinite values
df_sessions.replace([np.inf, -np.inf], np.nan, inplace=True)
df_sessions.fillna(0, inplace=True)

# Display time series features
df_sessions.head()
```

```python
# Simulate pitch conditions and player conflicts
def simulate_external_factors(n_players, n_weeks):
    external_data = []
    
    for player_id in range(n_players):
        for week in range(1, n_weeks + 1):
            # Pitch conditions (0-1, where 1 is worst)
            # Worse conditions in later weeks (winter months)
            base_pitch = 0.2 + (week / n_weeks) * 0.3
            pitch_condition = np.random.beta(2, 5) * base_pitch
            
            # Player conflicts (0-1, where 1 is most conflict)
            # More conflicts as season progresses
            base_conflict = 0.1 + (week / n_weeks) * 0.2
            conflict_level = np.random.beta(1, 10) * base_conflict
            
            # Fatigue accumulation over season
            fatigue_accumulation = week / n_weeks * 0.5
            
            external_data.append({
                'player_id': player_id,
                'week': week,
                'pitch_condition': pitch_condition,
                'conflict_level': conflict_level,
                'fatigue_accumulation': fatigue_accumulation
            })
    
    return pd.DataFrame(external_data)

# Generate external factors
external_df = simulate_external_factors(len(players), n_weeks)

# Merge with session data
df_sessions = df_sessions.merge(external_df, on=['player_id', 'week'])

# Add match intensity (higher on weekends)
def add_match_intensity(df):
    df['match_intensity'] = 0
    # Assume matches happen every 7 days (weekends)
    df.loc[df['week'] % 7 == 0, 'match_intensity'] = np.random.uniform(0.7, 1.0)
    return df

df_sessions = add_match_intensity(df_sessions)
```

```python
# 5. Injury Severity Classification (FIXED)

def classify_injury_severity(row, df_sessions):
    """
    Classify injury severity for a player
    
    Parameters:
    row: Player data row
    df_sessions: Time series data with ACWR information
    
    Returns:
    Injury severity classification
    """
    if row['Injury_Next_Season'] == 1:
        # Get ACWR data for this player
        player_id = row.name  # Use index as player_id
        player_acwr = df_sessions[df_sessions['player_id'] == player_id]['ACWR']
        
        # Use average ACWR if available, otherwise use a default
        avg_acwr = player_acwr.mean() if not player_acwr.empty else 1.0
        max_acwr = player_acwr.max() if not player_acwr.empty else 1.0
        
        # Calculate severity score (0-1)
        severity_score = (
            min(row['Previous_Injury_Count'] / 5, 1) * 0.25 +  # Up to 5 previous injuries
            (1 - min(row['Recovery_Score'] / 5, 1)) * 0.2 +    # Recovery score (0-5 scale)
            min(max(avg_acwr, 1.5) / 2, 1) * 0.2 +            # ACWR > 1.5 is risky
            row['Stress_Level_Score'] / 100 * 0.15 +           # Stress contribution
            (1 - min(row['Knee_Strength_Score'] / 100, 1)) * 0.1 +  # Weak knees
            (1 - min(row['Hamstring_Flexibility'] / 100, 1)) * 0.1  # Tight hamstrings
        )
        
        # Classify based on severity
        if severity_score > 0.7:
            return 'severe'  # Career-threatening (long recovery)
        elif severity_score > 0.4:
            return 'moderate'  # Season-ending
        else:
            return 'mild'  # Short recovery (weeks)
    return 'none'

# Apply severity classification - FIXED
df['injury_severity'] = df.apply(lambda x: classify_injury_severity(x, df_sessions), axis=1)

# Display severity distribution
severity_counts = df['injury_severity'].value_counts()
print("Injury Severity Distribution:")
print(severity_counts)
```

```python
# 6. Prepare Final Dataset for Modeling (FIXED)

# Aggregate time series data for each player
player_time_series = df_sessions.groupby('player_id').agg({
    'session_load': ['mean', 'std', 'max'],
    'ACWR': ['mean', 'max', 'last'],
    'load_spike': 'sum',
    'monotony': 'mean',
    'strain': 'mean',
    'pitch_condition': 'mean',
    'conflict_level': 'mean',
    'match_intensity': 'sum'
}).reset_index()

# Flatten column names
player_time_series.columns = ['_'.join(col).strip('_') for col in player_time_series.columns]
player_time_series.rename(columns={'player_id_': 'player_id'}, inplace=True)

# Merge with original data - FIXED to use proper player_id
df_model = df.merge(player_time_series, left_index=True, right_on='player_id', how='left')

# Create the target variable for severe injuries only
df_model['severe_injury'] = (df_model['injury_severity'] == 'severe').astype(int)

# Define features for modeling - UPDATED to include position if needed
features = [
    'Training_Hours_Per_Week', 'Stress_Level_Score', 'Workload_Stress_Index',
    'Sleep_Efficiency', 'Recovery_Score', 'Injury_Risk_Score', 'BMI',
    'Fitness_Workload_Ratio', 'Recovery_Deficit',
    'session_load_mean', 'session_load_std', 'session_load_max',
    'ACWR_mean', 'ACWR_max', 'ACWR_last',
    'load_spike_sum', 'monotony_mean', 'strain_mean',
    'pitch_condition_mean', 'conflict_level_mean', 'match_intensity_sum'
]

# For severe injury prediction
target_severe = 'severe_injury'

# For any injury prediction
target_any = 'Injury_Next_Season'

print(f"Final dataset shape: {df_model.shape}")
print(f"Number of features: {len(features)}")
print(f"Features: {features}")
```

```python
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from xgboost import XGBClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
import joblib
from imblearn.over_sampling import SMOTE
import numpy as np

# Prepare data
X = df_model[features]
y_any = df_model[target_any]

# Split data for ANY injury prediction
X_train_any, X_test_any, y_train_any, y_test_any = train_test_split(
    X, y_any, test_size=0.2, stratify=y_any, random_state=42
)

# Scale features for ANY injury prediction
scaler_any = StandardScaler()
X_train_scaled_any = scaler_any.fit_transform(X_train_any)
X_test_scaled_any = scaler_any.transform(X_test_any)

# Save the scaler for later use
joblib.dump(scaler_any, 'any_injury_scaler.pkl')


# Define models to try
models = {
    'Random Forest': RandomForestClassifier(random_state=42),
    'Gradient Boosting': GradientBoostingClassifier(random_state=42),
    'XGBoost': XGBClassifier(random_state=42, eval_metric='logloss'),
    'Logistic Regression': LogisticRegression(random_state=42, max_iter=1000)
}

# Evaluate models for any injury prediction
print("üîç Evaluating models for ANY injury prediction:")
print("=" * 50)

any_injury_results = {}
for name, model in models.items():
    model.fit(X_train_scaled_any, y_train_any)
    y_pred = model.predict(X_test_scaled_any)
    y_prob = model.predict_proba(X_test_scaled_any)[:, 1]

    # Store results
    any_injury_results[name] = {
        'model': model,
        'report': classification_report(y_test_any, y_pred, output_dict=True),
        'auc': roc_auc_score(y_test_any, y_prob)
    }

    print(f"\n{name}:")
    print(f"AUC: {any_injury_results[name]['auc']:.3f}")
    print(classification_report(y_test_any, y_pred))

# Select best model for any injury prediction
best_any_model = max(any_injury_results.items(), key=lambda x: x[1]['auc'])

print(f"\nüéØ Best model for ANY injuries: {best_any_model[0]} (AUC: {best_any_model[1]['auc']:.3f})")

# Save the best model
joblib.dump(best_any_model[1]['model'], 'any_injury_model.pkl')
```

```python
def predict_injury_risk(player_data, model_type='any'):
    """
    Predict injury risk for a new player
    
    Parameters:
    player_data: Dictionary or DataFrame with player features
    model_type: 'any' for any injury, 'severe' for severe injuries only
    
    Returns:
    Dictionary with prediction results
    """
    # Load appropriate model and scaler
    if model_type == 'any':
        model = joblib.load('tuned_injury_model.pkl')
    else:
        model = joblib.load('severe_injury_model.pkl')
    
    scaler = joblib.load('injury_scaler.pkl')
    
    # Convert to DataFrame if needed
    if isinstance(player_data, dict):
        player_df = pd.DataFrame([player_data])
    else:
        player_df = player_data.copy()
    
    # Ensure all required features are present
    for feature in features:
        if feature not in player_df.columns:
            player_df[feature] = 0  # Default value if missing
    
    # Select and scale features
    X_new = player_df[features]
    X_new_scaled = scaler.transform(X_new)
    
    # Make prediction
    prediction = model.predict(X_new_scaled)
    probability = model.predict_proba(X_new_scaled)
    
    # Prepare results
    result = {
        'prediction': int(prediction[0]),
        'probability': float(probability[0][1]),
        'risk_level': 'high' if probability[0][1] > 0.7 else 'medium' if probability[0][1] > 0.3 else 'low'
    }
    
    return result

# Example usage
example_player = {
    'Training_Hours_Per_Week': 12.5,
    'Stress_Level_Score': 45.0,
    'Workload_Stress_Index': 12.5 * 45.0,
    'Sleep_Efficiency': 8.5 / 12.5,
    'Recovery_Score': 3.8,
    'Injury_Risk_Score': -4.2,
    'BMI': 22.5,
    'Fitness_Workload_Ratio': (75 + 80) / 12.5,
    'Recovery_Deficit': 45 - 3.8*10,  # Approximate conversion
    'session_load_mean': 11.8,
    'session_load_std': 1.2,
    'session_load_max': 14.5,
    'ACWR_mean': 1.1,
    'ACWR_max': 1.4,
    'ACWR_last': 1.2,
    'load_spike_sum': 2,
    'monotony_mean': 1.8,
    'strain_mean': 21.2,
    'pitch_condition_mean': 0.3,
    'conflict_level_mean': 0.1,
    'match_intensity_sum': 4
}

prediction = predict_injury_risk(example_player)
print(f"Example Player Prediction: {prediction}")
```
